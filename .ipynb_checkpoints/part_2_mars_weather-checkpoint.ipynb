```python
# Import relevant libraries
from splinter import Browser
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
browser = Browser('chrome')
```

### Step 1: Visit the Website

Use automated browsing to visit the [Mars Temperature Data Site](https://static.bc-edx.com/data/web/mars_facts/temperature.html). Inspect the page to identify which elements to scrape.

   > **Hint** To identify which elements to scrape, you might want to inspect the page by using Chrome DevTools to discover whether the table contains usable classes.


```python
# Visit the website
# https://static.bc-edx.com/data/web/mars_facts/temperature.html
url = "https://static.bc-edx.com/data/web/mars_facts/temperature.html"
browser.visit(url)
```

### Step 2: Scrape the Table

Create a Beautiful Soup object and use it to scrape the data in the HTML table.

Note that this can also be achieved by using the Pandas `read_html` function. However, use Beautiful Soup here to continue sharpening your web scraping skills.

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, "html.parser")
```

```python
# Extract all rows of data
weather_table = soup.find("table")  
df = pd.read_html(str(weather_table))[0]
```

### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Here’s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create an empty list
mars_weather_data = []

# Loop through the scraped data to create a list of rows
table = soup.find("table")

# Loop through all table rows (skip the header)
for row in table.find_all("tr")[1:]:
    # Extract each cell (td) in the row
    cells = row.find_all("td")
    
    # Append row data as a list to mars_weather_data
    mars_weather_data.append([
        int(cells[0].text.strip()),  # id
        cells[1].text.strip(),       # terrestrial_date
        int(cells[2].text.strip()),  # sol
        float(cells[3].text.strip()), # ls
        int(cells[4].text.strip()),  # month
        float(cells[5].text.strip()), # min_temp
        float(cells[6].text.strip())  # pressure
    ])
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
# Define column names
columns = ['id', 'terrestrial_date', 'sol', 'ls', 'month', 'min_temp', 'pressure']

# Create DataFrame
df = pd.DataFrame(mars_weather_data, columns=columns)

# Convert terrestrial_date to datetime format
df['terrestrial_date'] = pd.to_datetime(df['terrestrial_date'])

# Display the DataFrame

print(df.head())
```

```python
# Confirm DataFrame was created successfully
df.info(), df.head()
```

### Step 5: Analyze the Data

Analyze your dataset by using Pandas functions to answer the following questions:

1. How many months exist on Mars?
2. How many Martian (and not Earth) days worth of data exist in the scraped dataset?
3. What are the coldest and the warmest months on Mars (at the location of Curiosity)? To answer this question:
    * Find the average the minimum daily temperature for all of the months.
    * Plot the results as a bar chart.
4. Which months have the lowest and the highest atmospheric pressure on Mars? To answer this question:
    * Find the average the daily atmospheric pressure of all the months.
    * Plot the results as a bar chart.
5. About how many terrestrial (Earth) days exist in a Martian year? To answer this question:
    * Consider how many days elapse on Earth in the time that Mars circles the Sun once.
    * Visually estimate the result by plotting the daily minimum temperature.


```python
# 1. How many months are there on Mars?
# Count the unique Martian months
num_martian_months = df['month'].nunique()

# Display result
print(f"There are {num_martian_months} months on Mars.")
```

```python
# 2. How many sols (Martian days) worth of data are there?
num_sols = df['sol'].nunique()
print(f"There are {num_sols} Martian sols worth of data.")
```

```python
# 3. What is the average minimum temperature by month?
# Calculate the average minimum temperature for each Martian month
avg_min_temp_by_month = df.groupby('month')['min_temp'].mean()

# Display the result
print(avg_min_temp_by_month)
```

```python
# Plot the average minimum temperature by month
# Plot the data
avg_min_temp_by_month.plot(kind='bar', color='blue', figsize=(10,5))

# Add labels and title
plt.xlabel("Martian Month")
plt.ylabel("Average Minimum Temperature (°C)")
plt.title("Average Minimum Temperature by Martian Month")
plt.grid(axis='y')
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location by sorting the previous graph
# Calculate the average minimum temperature for each Martian month
avg_min_temp_by_month = df.groupby('month')['min_temp'].mean()

# Sort the months by temperature (ascending order)
sorted_temp_by_month = avg_min_temp_by_month.sort_values()

# Plot the sorted data
plt.figure(figsize=(10,5))
sorted_temp_by_month.plot(kind='bar', color='red')

# Add labels and title
plt.xlabel("Martian Month")
plt.ylabel("Average Minimum Temperature (°C)")
plt.title("Coldest and Hottest Months at Curiosity's Location")
plt.grid(axis='y')

# Show the plot
plt.show()

# Identify the coldest and hottest months
coldest_month = sorted_temp_by_month.idxmin()
hottest_month = sorted_temp_by_month.idxmax()

# Display results
print(f"Coldest Month: {coldest_month} (Lowest Avg Temp: {sorted_temp_by_month.min():.2f}°C)")
print(f"Hottest Month: {hottest_month} (Highest Avg Temp: {sorted_temp_by_month.max():.2f}°C)")
```

```python
# 4. What is the average pressure by month?
# Calculate the average atmospheric pressure for each Martian month
avg_pressure_by_month = df.groupby('month')['pressure'].mean()

# Display the result
print(avg_pressure_by_month)
```

```python
# Plot the average pressure by month
# Calculate the average atmospheric pressure for each Martian month
avg_pressure_by_month = df.groupby('month')['pressure'].mean()

# Plot the data
plt.figure(figsize=(10,5))
avg_pressure_by_month.plot(kind='bar', color='purple')

# Add labels and title
plt.xlabel("Martian Month")
plt.ylabel("Average Pressure (Pa)")
plt.title("Average Atmospheric Pressure by Martian Month")
plt.grid(axis='y')

# Show the plot
plt.show()
```

```python
# Identify the lowest and highest pressure months in Curiosity's location by sorting the previous graph
# Calculate the average atmospheric pressure for each Martian month
avg_pressure_by_month = df.groupby('month')['pressure'].mean()

# Sort the months by pressure (ascending order)
sorted_pressure_by_month = avg_pressure_by_month.sort_values()

# Plot the sorted data
plt.figure(figsize=(10,5))
sorted_pressure_by_month.plot(kind='bar', color='green')

# Add labels and title
plt.xlabel("Martian Month")
plt.ylabel("Average Pressure (Pa)")
plt.title("Lowest and Highest Pressure Months at Curiosity's Location")
plt.grid(axis='y')

# Show the plot
plt.show()

# Identify the lowest and highest pressure months
lowest_pressure_month = sorted_pressure_by_month.idxmin()
highest_pressure_month = sorted_pressure_by_month.idxmax()

# Display results
print(f"Lowest Pressure Month: {lowest_pressure_month} (Min Avg Pressure: {sorted_pressure_by_month.min():.2f} Pa)")
print(f"Highest Pressure Month: {highest_pressure_month} (Max Avg Pressure: {sorted_pressure_by_month.max():.2f} Pa)")
```

```python
# How many terrestrial (Earth) days are there in a Martian year?
# Visually estimate the result by plotting the daily minimum temperature of each observation in the data set.
# Plot daily minimum temperature over time
plt.figure(figsize=(12,6))
plt.plot(df['terrestrial_date'], df['min_temp'], marker='o', linestyle='-', markersize=2, color='b')

# Add labels and title
plt.xlabel("Terrestrial Date")
plt.ylabel("Minimum Temperature (°C)")
plt.title("Daily Minimum Temperature on Mars Over Time")
plt.xticks(rotation=45)
plt.grid(True)

# Show the plot
plt.show()

```
